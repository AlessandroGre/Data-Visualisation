## Plotting Methods

The Plotting Methods.ipynb notebook contains a series of methods designed for visualizing financial data. These methods leverage popular Python libraries like Matplotlib and Plotly to create various types of plots, including line charts, bar charts, and candlestick charts. This notebook is intended to aid analysts, traders, and researchers in gaining insights from financial time series and other related data through effective and clear visual representations.

```python
import yfinance as yf
import mplfinance as mpf
import warnings
warnings.filterwarnings('ignore')
```

<font color= #6B6B6B> <font size = 5> Import stock prices </font> <a name="Project_Packages"></a>
<hr style="border:0.01in solid gray"> </hr>

```python
# Find the financial ticker for the chosen company
ticker = 'TSLA'

# Set start and end date of the period of interest
start_date = '2023-01-01'
end_date = '2023-12-31'

# Import stock prices from Yahoo Finance
prices = yf.download(ticker, start=start_date, end=end_date)
prices
```

<hr style="border:0.01in solid gray"> </hr>
<font color= #6B6B6B> <font size = 5> Visualize a simple OHLC plot </font> <a name="Project_Packages"></a>
<hr style="border:0.01in solid gray"> </h

```python
# Create a custom style using the custom market colors
custom_style = mpf.make_mpf_style(facecolor='w')

# Plot with custom candle colors
mpf.plot(prices, type='ohlc', figsize=(20,8), style=custom_style)
```

<hr style="border:0.01in solid gray"> </hr>
<font color= #6B6B6B> <font size = 5> Visualize a candlestick pattern </font> <a name="Project_Packages"></a>
<hr style="border:0.01in solid gray"> </h

```python
# Create a custom style using the custom market colors
custom_style = mpf.make_mpf_style(facecolor='w')

# Plot with custom candle colors
mpf.plot(prices, type="candle", figsize=(20,8), style=custom_style)
```

```python
# Define custom market colors for bullish and bearish candles
custom_marketcolors = mpf.make_marketcolors(up='g', down='r')

# Create a custom style using the custom market colors
custom_style = mpf.make_mpf_style(marketcolors=custom_marketcolors, facecolor='w')

# Plot with custom candle colors
mpf.plot(prices, type="candle", figsize=(20,8), style=custom_style)
```

```python
# Define custom market colors for bullish and bearish candles
custom_marketcolors = mpf.make_marketcolors(up='g', down='r')

# Create a custom style using the custom market colors
custom_style = mpf.make_mpf_style(marketcolors=custom_marketcolors, facecolor='w')

# Plot with custom candle colors
mpf.plot(prices, type="candle", figsize=(20,8), style=custom_style, volume=True)
```

<font color= #6B6B6B> <font size = 5> Intraday charting </font> <a name="Project_Packages"></a>

```python
intraday = yf.download(ticker, period="5d", interval="1m")
iday = intraday.iloc[-100:, :]
mpf.plot(iday, type="candle", figsize=(20,8), mav=(7, 12))
```
